**What are the common characteristics of individuals with high stress levels?**

```python
bins = [0, 3, 6, 10]
labels = ['Low', 'Medium', 'High']

df['Stress Level Category'] = pd.cut(df['Stress Level'], bins=bins, labels=labels)
plt.figure(figsize=(5,5))
plt.scatter(x='Stress Level Category',y='Heart Rate',data=df)
plt.title('Heart Rate by Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Heart Rate')
plt.tight_layout()
plt.show()
plt.figure(figsize=(5,5))
plt.scatter(x='Stress Level Category',y='Blood Pressure',data=df)
plt.title('Blood Pressure by Stress Level')
plt.xlabel('Stress Level ')
plt.ylabel('Blood Pressure')
plt.tight_layout()
plt.show()
plt.figure(figsize=(5,5))
plt.scatter(x='Stress Level Category',y='Cholesterol',data=df)
plt.title('Cholesterol by Stress Level')
plt.xlabel('Stress Level ')
plt.ylabel('Cholesterol')
plt.tight_layout()
plt.show()
```

**How does obesity relate to other health indicators like diabetes and
cholesterol?**


While obesity is a significant risk factor, it's not the only cause. Other factors like family history and diet also play a role

```python
# Scatter plots for Cholesterol vs. Obesity and Diabetes
plt.figure(figsize=(12, 6))  # Adjust for better visualization

# Cholesterol vs. Obesity
plt.subplot(1, 2, 1)  # Create a subplot for the first plot
sns.scatterplot(x="Obesity", y="Cholesterol", data=df)
plt.title("Cholesterol vs. Obesity")
plt.xlabel("Obesity (Yes/No)")
plt.ylabel("Cholesterol")

# Cholesterol vs. Diabetes
plt.subplot(1, 2, 2)  # Create a subplot for the second plot
sns.scatterplot(x="Diabetes", y="Cholesterol", data=df)
plt.title("Cholesterol vs. Diabetes")
plt.xlabel("Diabetes (Yes/No)")
plt.ylabel("Cholesterol")
plt.show()
```

**Is there a pattern between exercise-induced angina and heart disease?**

analysis and consulting a medical professional are essential for drawing definitive conclusions

```python
# Create a contingency table for Exercise Induced Angina by Heart Disease
angina_by_heart_disease = pd.crosstab(
    df['Heart Disease'], df['Exercise Induced Angina']
)
# Create a bar plot
plt.figure(figsize=(4, 4))
sns.barplot(x=angina_by_heart_disease.index, y=angina_by_heart_disease.sum())
plt.xlabel("Heart Disease")
plt.ylabel("Count (Exercise Induced Angina)")
plt.title("Exercise Induced Angina by Heart Disease")
plt.show()
```

**What types of chest pain are most common among individuals with heart
